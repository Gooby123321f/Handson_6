## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:
```powershell
Monthly Sales program

COMMAND MENU
view	- View sales for specified month 
edit	- Edit sales for specified month 
totals - View sales summary for year 
exit	- Exit program

Command: view
Three-letter Month: jan
Sales amount for Jan is 14,317.00.

Command: edit
Three-letter Month: jan 
Sales Amount: 15293
Sales amount for Jan is 15,293.00.

Command: totals
Yearly total: 67,855.00
Monthly average: 5,654.58

Command: view
Three-letter Month: july 
Invalid three-letter month.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
def displayMenu():
    print("COMMAND MENU\nview - View sales for specified month\nedit - Edit sales for specified month"
    "\ntotal - View sales summary for whole year\nexit - Exit program\n")
    
def view(dictionary, monthsArr):
    month = input("Three-Letter Month: ")
    boolean = True
    while(boolean):
        for i in range(0, len(monthsArr), 1):                
            if(month.lower() == monthsArr[i].lower()):
                boolean = False
                break
            elif(i == len(monthsArr)-1 and not month.lower() == monthsArr[i].lower()):
                month = input("Invalid month input! enter a valid 3 letter abbreviation for a month: ")
                i = 0
        
    print("Sales amount for " + month.capitalize() + ": " + str(round(dictionary[month.capitalize()], 2)))

def edit(dictionary, monthsArr):
    month = input("Three-Letter Month: ")
    boolean = True
    while(boolean):
        for i in range(0, len(monthsArr), 1):                
            if(month.lower() == monthsArr[i].lower()):
                boolean = False
                break
            elif(i == len(monthsArr)-1 and not month.lower() == monthsArr[i].lower()):
                month = input("Invalid month input! enter a valid 3 letter abbreviation for a month: ")
                i = 0
        
    newMoney = float(input("Sales Amount: "))
    dictionary[month.capitalize()] = newMoney
    print("Sales amount for " + month.capitalize() + " is " + str(newMoney))
    
    with open("monthly_sales.txt", "w") as f:
        for key in dictionary:
            f.write(key + "\t" + str(dictionary[key]) + "\n")

def total(dictionary):
    totalMoney = 0.0
    for key in dictionary:
        totalMoney += dictionary[key]
    print("Yearly Total: " + str(round(totalMoney, 2)))
    print("Monthly Average: " + str(round(totalMoney / 12, 2)))
        
    
def main():
    ##declares an array of months for validation and the sales dictionary
    monthArray = []
    sales = {}
    with open("monthly_sales.txt", "r") as f:
        for line in f:
            temp = line.split("\t")
            monthArray.append(temp[0])
            sales[temp[0]] = float(temp[1].replace("\n", ""))
            
    displayMenu()
    while(True):
        command = input("Command: ")
        while not(command == "view" or command == "edit" or command == "total" or command == "exit"):
            command = input("\nInvalid input! Enter a command from the list above: ")
        if(command == "exit"):
            break
        elif(command == "view"):
            view(sales, monthArray)
        elif(command == "edit"):
            edit(sales, monthArray)
        else:
            total(sales)
        print("\n")
    print("Bye!")
            
if __name__ == "__main__":
    main()
        
````
